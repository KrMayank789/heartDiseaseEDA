```python

import numpy as np 
import pandas as pd 
import seaborn as sns



import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
        
df = pd.read_csv('/kaggle/input/heart-csv/heart.csv')

df.head()
```
